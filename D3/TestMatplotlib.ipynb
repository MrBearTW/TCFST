```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

np.random.seed()    # ()內給值的話會跑出一樣的結果，沒給就會隨機
x = pd.period_range(pd.datetime.now(),periods=200,freq='d')    #從當前時間開始取200天
x = x.to_timestamp().to_pydatetime()    # 把2017-11-06轉換成 ==>(2017, 11, 6, 0, 0)
y = np.random.randn(200,3).cumsum(0)    # 建議3個擁有200個元素的隨機陣列    # cumsum(0)從第0軸開始加總
#print(x)
#print(y)
plots =plt.plot(x,y)
```

```python
plots = plt.plot(x, y, label='')
plt.gcf().set_size_inches(8, 4)    # 圖的尺寸為 8 * 4 吋
plt.legend(plots, ('foo', 'bar', 'baz')    # 3條線標籤
           ,loc='best'    # 避免圖表被遮住
           ,framealpha=0.25    # 設定圖例透明度
           ,prop={'size':'small', 'family':'monospace'})    # 調整圖例字型
plt.title('Random trends')
plt.xlabel('Date')
plt.ylabel('Cum. sum')
plt.grid(True)    # 格線開啟
plt.figtext(0.9, 0.001    # 文字的位置，相對於圖的尺寸(原點應該是在左下角)
            ,u'© Acme Designs 2015'   # 內容
            ,ha='right', va='bottom')    # 位置
plt.tight_layout()    # 確保所有元素皆位於方盒內
#plt.savefig('mpl_3lines_custom.svg')    # 存檔 PNG PDF PS EPS SVG都可以
```

```python
fig = plt.figure(figsize=(8,4))
#--- Main Axes
ax = fig.add_axes((0.1,0.1,0.8,0.8))
ax.set_title('Main Axes with Insert Child Axes')
ax.plot(x, y[:,2])   # 取第3組數據
ax.set_xlabel('Date')
ax.set_ylabel('Cum. sum')
#--- Inserted Axes
ax = fig.add_axes([0.15,0.15,0.3,0.3])    # bottom left top right
ax.plot(x, y[:,1], color='g')    # 取第2組數據
ax.set_xticks([]);    # 移除嵌入圖表的x標籤與標號
```

```python
fig, axes = plt.subplots(nrows=3,ncols=1,    # 3列(直) 1欄
                         sharex=True,sharey=True,    # 共享 x y軸
                         figsize=(8,8))
labelled_data = zip(y.transpose(), ('foo', 'bar', 'baz'), ('b', 'g', '#111111'))
fig.suptitle('Three Random Trends', fontsize=16)
for i, ld in enumerate(labelled_data):
    ax = axes[i]
    ax.plot(x, ld[0], label=ld[1], color=ld[2])
    ax.set_ylabel('Cum. sum')
    ax.legend(loc='upper left', framealpha=0.5, prop={'size':'small'})
axes[-1].set_xlabel('Date')    # 共享x軸的最後一個
```

```python

labels = ["Physics", "Chemistry", "Literature", "Peace"]
data =   [3, 6, 10, 4]

bar_width = 0.5
xlocations = np.array(range(len(data)))+0.5
plt.bar(xlocations, data, width=bar_width)
plt.yticks(range(0, 12))
plt.xticks(xlocations+ bar_width/2 , labels)    # 此處放上標籤
plt.xlim(0, xlocations[-1]+bar_width*2)    # 設定X上下限，允許組又空白填補一個bar_width的寬度
plt.title("Prizes won by Fooland")
plt.gca().get_xaxis().tick_bottom()
plt.gca().get_yaxis().tick_left()
plt.gcf().set_size_inches((8,4))
```

```python
np.random.seed()
num_points = 100
gradient = 0.5
x = np.array(range(num_points))
y = np.random.randn(num_points) * 10 + x*gradient    # randn=>隨機給出一個0~10之間的數
fig, ax = plt.subplots(figsize=(8, 4))
ax.scatter(x, y)
m,c=np.polyfit(x,y,1)    # 1代表一維，m斜率，c常數
ax.plot(x,m*x+c)    # 畫出那一條線
fig.suptitle('A Simple Scatterplot')
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
# FacetGrid
g=sns.FacetGrid(tips,col="sex")    # 區別的
g.map(plt.scatter,"total_bill","tip")    # x y 軸分別是
```

```python
pal = dict(Female='red', Male='blue')
g = sns.FacetGrid(tips, col="smoker", hue="sex", 
                  palette=pal, size=4, aspect=1, 
                  hue_kws={"marker": ["D", "s"]})    # D鑽石 s方型
g.map(plt.scatter, "total_bill", "tip", alpha=.4)
g.add_legend();
```

```python
# Pairgrid
iris=sns.load_dataset('iris')
sns.set(font_scale=1.5)    # 修改字型大小
g=sns.PairGrid(iris,hue="species")    # 根據品種設定不同顏色
g.map_diag(plt.hist)    # 品種為度的柱狀圖，放在格子對角線上
g.map_offdiag(plt.scatter)    # 使用標準散布圖比較對角線的各維度
g.add_legend();
```
