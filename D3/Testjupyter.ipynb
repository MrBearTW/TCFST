```python
#------born_in clean------
import numpy as np
#df.born_in.count()    #沒替代之前是1052
df.born_in.replace('',np.nan,inplace=True)    # ' ' 用np套件的NaN取代
#df.born_in.count()    # 替代之後得真實數量是142
#df.born_in
df.born_in.describe()
```

```python
#------name clean------
#df[df.name.str.contains('\*')]['name']    #看一下姓名後面有*的
df.name = df.name.str.replace('*','')    # 消除 * 符號
df.name
```

```python
df.name = df.name.str.strip()    # 消除 部分名字後面出現空格
df.name
```

```python
np.nan == np.nan # 根據定義：NaN的布林運算是一定會傳回False
```

```python
#df    # 再看一下表格
df = df[df.born_in.isnull()]    # 用isnull查出哪一些是born_in是NaN
#df.count()
df = df.drop('born_in', axis=1)  #丟掉born_in這一欄(直)    # 預設axis=0列(橫)  1欄(直)
df.head()
```
