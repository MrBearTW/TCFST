```python
#------born_in clean------
import numpy as np
df.born_in.replace('',np.nan,inplace=True)
df.born_in
#df.born_in.describe()
```

```python
#------name clean------
df.name = df.name.str.replace('*','')    # 消除 * 符號
df.name
```

```python
df.name = df.name.str.strip()    # 消除 部分名字後面出現空格
df.name
```

```python
df_drop_born_in =df[df.name.isnull()]    #查出哪一些是born_in是NaN
df_drop_born_in = df.drop('born_in', axis=1)
df_drop_born_in.head()
```

```python
df_drop_born_in_drop = df_drop_born_in.drop('born_in',axis=1)    # 拿掉born_in是NaN的那一直列
df_drop_born_in_drop.head()
df = df_drop_born_in_drop
```

```python
dupse_by_name= df[df.duplicated('name')]
dupse_by_name.count()
```

```python
all_dupes = df[df.duplicated('name')|df.duplicated('name',keep='last')]    # keep='last'  >>> 重複出現兩次以上的也都計算
all_dupes.count()
```
